```python
#import relevant libraries 
import numpy as np
import matplotlib.pyplot as plt

## Define Domain
L=2*np.pi
N=64
dx=L/N
x = np.arange(0,L,dx)

## Define Signal
y = np.sin(x)

## Derivative computed analytically 
dy=np.cos(x)
```

```python
#Plot Signal and analytic derivative
plt.plot(x,y,color='black', label='Signal')
plt.plot(x,dy, color='orange', label='Analytic Derivative')
plt.legend(bbox_to_anchor=(1, 1))
plt.title("Signal and Derivative")
plt.show()

```

```python
##Compute Derivative using finite differences and matricies

M=np.zeros((N,N)) #define zero matrix

for i in range(N-1): #construct finite differences matrix
    M[i,i]=-1
    M[i,i+1]=1
M[-1,-1]=1
M[N-1,-2]=-1


dyFDM=(np.matmul(M,y))/dx #Calculate the finite differnce derivative approximation
```

```python
plt.plot(x,dyFDM,label="Approximate Derivative using Finite Differences")
plt.plot(x,dy,label="Analytic Derivative")
plt.legend(bbox_to_anchor=(1, 1))
plt.title("Derivatives and Matricies")
plt.show()
```

```python
## Approximate derivative using finite differences and a for loop to check your work

dyFD=np.zeros(len(dy),dtype='float')
for i in range(len(dy)-1):
    dyFD[i]=(y[i+1]-y[i])/dx
    
dyFD[-1]=dyFD[-2]
plt.plot(x,dyFD,label="Approximate Derivative using Finite Differences")
plt.plot(x,dy,label="Analytic Derivative")
plt.legend(bbox_to_anchor=(1, 1))
plt.title("Derivatives and Matricies")
plt.show()
```
